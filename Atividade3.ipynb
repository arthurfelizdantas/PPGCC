```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Criando uma população artificial com distribuição Gaussiana
np.random.seed(42)
populacao = np.random.normal(loc=50, scale=10, size=10000)  # média = 50, desvio padrão = 10, tamanho = 10000

# Parâmetros da população
media_pop = np.mean(populacao)
desvio_pop = np.std(populacao)

# Amostra simples
amostra = np.random.choice(populacao, size=100, replace=False)
media_amostra = np.mean(amostra)

# Distribuição de amostragem (sampling distribution)
n_amostras = 1000
medias_amostras = [np.mean(np.random.choice(populacao, size=100, replace=False)) for _ in range(n_amostras)]

# Avaliações de probabilidade de possíveis eventos
# Probabilidade de uma amostra ter média maior que 52
prob_maior_52 = 1 - norm.cdf(52, loc=media_pop, scale=desvio_pop / np.sqrt(100))

# Plotando os gráficos
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# População
axs[0].hist(populacao, bins=50, density=True, alpha=0.6)
x = np.linspace(20, 80, 1000)
axs[0].plot(x, norm.pdf(x, media_pop, desvio_pop), 'r--')
axs[0].set_title('Distribuição da População')

# Distribuição da amostra única
axs[1].hist(amostra, bins=20, density=True, alpha=0.6)
axs[1].plot(x, norm.pdf(x, media_amostra, np.std(amostra)), 'g--')
axs[1].set_title('Distribuição da Amostra Única')

# Distribuição das médias das amostras (Sampling Distribution)
axs[2].hist(medias_amostras, bins=30, density=True, alpha=0.6)
axs[2].plot(x, norm.pdf(x, media_pop, desvio_pop / np.sqrt(100)), 'b--')
axs[2].set_title('Distribuição de Amostragem (Sampling Distribution)')

plt.tight_layout()
plt.show()

(media_pop, desvio_pop, media_amostra, prob_maior_52)
```
